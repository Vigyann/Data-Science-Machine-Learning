```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix,classification_report,recall_score,accuracy_score,roc_auc_score,precision_score,roc_curve,precision_recall_curve

le_type = LabelEncoder()
le_region = LabelEncoder()
le_status = LabelEncoder()
le_age=LabelEncoder()

df_f['type_encoded'] = le_type.fit_transform(df_f['type'])
df_f['region_encoded'] = le_region.fit_transform(df_f['region'])
df_f['status_encoded'] = le_status.fit_transform(df_f['status'])
df_f['age_encoded'] = le_age.fit_transform(df_f['age'])
```

```python
df_f = df_f.drop(['Longitude'], axis=1)
```

```python
df_f.head()
```

```python
features = ['bhk', 'type_encoded', 'region_encoded', 
           'price','age_encoded', 'Latitude','area' ]
x = df_f[features]
y = df_f['status_encoded']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)
```

```python
ss = StandardScaler()
x_train_scaled = ss.fit_transform(x_train)
x_test_scaled = ss.transform(x_test)

model = LogisticRegression()   
model.fit(x_train_scaled, y_train)

y_pred = model.predict(x_test_scaled)
```

```python

cl_report = classification_report(y_test, y_pred)

recallscore = recall_score(y_test, y_pred)
accuracyscore= accuracy_score(y_test, y_pred)
c_matrxix= confusion_matrix(y_test, y_pred)

print("\n Performance Metrics\n:")
print(f"classification report: {cl_report} ")
print(f"recall score: {recallscore:.2f} ")
print(f"accuracy score: {accuracyscore:.2f} ")
print(f"confusion matrix\n : {c_matrxix}")
```

```python
import pandas as pd
import numpy as np

def predict_status(bhk, price, area, property_type, region, age, latitude):

    type_encoded = le_type.transform([property_type])[0]
    region_encoded = le_region.transform([region])[0]
    age_encoded = le_age.transform([age])[0]

    input_data = pd.DataFrame({
        'bhk': [bhk],
        'area': [area],
        'type_encoded': [type_encoded],
        'region_encoded': [region_encoded],
        'age_encoded': [age_encoded],
        'Latitude': [latitude],
        'price': [price]
    })

   
    input_data = input_data[features]

  
    input_scaled = ss.transform(input_data)

   
    prediction = model.predict(input_scaled)[0]  # 0 or 1 for status
    prediction_proba = model.predict_proba(input_scaled)[0][1]  # Probability for the positive class

 
    status = "Under Construction" if prediction == 1 else "Ready to move"
    
    return status, prediction_proba

# Sample usage of the function with new data
sample_status, sample_probability = predict_status(
    bhk=2,
    price=0.41,
    area=950,
    property_type='Apartment',
    region='khopoli',
    age='New',
    latitude=18.78775
)

print(f"\nPrediction for 2 BHK Apartment in Thane West having price 0.60 Cr of New type is :")
print(f"Status: {sample_status}")
print(f"Probability of Sale: {sample_probability:.2f}")
```
