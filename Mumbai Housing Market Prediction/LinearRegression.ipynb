```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

le_type = LabelEncoder()
le_region = LabelEncoder()
le_status = LabelEncoder()
le_age=LabelEncoder()

df_f['type_encoded'] = le_type.fit_transform(df_f['type'])
df_f['region_encoded'] = le_region.fit_transform(df_f['region'])
df_f['status_encoded'] = le_status.fit_transform(df_f['status'])
df_f['age_encoded'] = le_age.fit_transform(df_f['age'])
```

```python
df_f = df_f.drop(['Longitude'], axis=1)
```

```python
df_f.head()
```

```python
features = ['bhk', 'type_encoded','area', 'region_encoded', 'status_encoded','age_encoded', 'Latitude' ]
x = df_f[features]
y = df_f['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=42)
```

```python
ss = StandardScaler()
x_train_scaled = ss.fit_transform(x_train)
x_test_scaled = ss.transform(x_test)

model = LinearRegression()   
model.fit(x_train_scaled, y_train)

y_pred = model.predict(x_test_scaled)
```

```python
import numpy as np
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\n Performance Metrics:")
print(f"MSE: {mse:.2f} ")
print(f"RMSE: {rmse:.2f} ")
print(f"MAE: {mae:.2f} ")
print(f"R-squared : {r2:.4f}")
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs. Predicted Prices')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(y_pred, y_test - y_pred)
plt.xlabel('Predicted Price')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.hlines(y=0, xmin=y_pred.min(), xmax=y_pred.max(), color='red', lw=2)
plt.show()
```
